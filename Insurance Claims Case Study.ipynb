Q16. Is there any similarity in the amount claimed by males and females?

```python
#Q16
claims_male = customer_claims["claim_amount"].loc[customer_claims["gender"]=="Male"]
claims_female = customer_claims["claim_amount"].loc[customer_claims["gender"]=="Female"]

```

```python
claims_male.mean()
```

```python
claims_female.mean()
```

```python
import scipy.stats as stats

equal_variance = stats.ttest_ind(a= claims_male,
                b= claims_female,
                equal_var=True)    
equal_variance.statistic
```

```python

unequal_variance = stats.ttest_ind(a= claims_male,
                b= claims_female,
                equal_var=False)    
unequal_variance.statistic
```

```python
unequal_variance.statistic - equal_variance.statistic
```

```python
p_value = equal_variance.pvalue
p_value
```

```python
if (p_value<0.05):
    print("We reject null hypothesis")
else:
    print("We fail to reject null hypothesis/there is some relationship between the two")
```

Q17. Is there any relationship between age category and segment?

```python
#Q17
age_segment_relation=pd.crosstab(customer_claims["Age_Group"],customer_claims["Segment"],margins=True)
```

```python
age_segment_relation
```

```python
chi_test=stats.chi2_contingency(observed= age_segment_relation)
chi_test
```

```python
chi_test_stats=chi_test[0]
chi_test_stats
```

```python
p_test=chi_test[1]
p_test
```

```python
#p value is greater than 0.05, we fail reject the null hypothesis. 
#Therefore there is no relationship between age category and segment
```

19. Is there any difference between age groups and insurance claims?

#Q19

H0 : (No difference between age groups and insurance claims or No influence of age groups on insurance claims)


Ha : (There is some difference between age groups and insurance claims or there is some influence of age groups on insurance claims)

```python
#converting age-group into categorical variable
customer_claims["Age_Group_cat"]=np.where((customer_claims.Age.astype(int)<18),1,np.where(((customer_claims.Age.astype(int)>18) & (customer_claims.Age.astype(int)<30)),2,
                                     np.where(((customer_claims.Age.astype(int)>=30) & (customer_claims.Age.astype(int)<60)),3,4)))
```

```python
customer_claims
```

```python
customer_claims["total_policy_claims"].value_counts()
```

```python
ct=pd.crosstab(customer_claims.total_policy_claims,customer_claims.Age_Group_cat,margins=True)
ct
```

```python
stats.chi2_contingency(ct)
```

Since the p value of the test is greater than 0.05, we fail reject the null hypothesis. Therefore, there is no difference between age groups and insurance claims or No influence of age groups on insurance claims
